```python
import geopandas as gpd
# mengimpor modul geopandas dan memberikan alias atau nama singkat "gpd" yang akan digunakan untuk mengakses fungsi, kelas, dan atribut yang ada di dalam modul
```

```python
gdf = gpd.read_file("RBI25K_BANGUNAN_PT.json")
gdf.head()
# gdf.info()
# Pada tahap ini membaca data dari file dan menyimpanya pada gdf 
```

```python
gdf = gdf[[ 'REMARK', 'LCODE', 'geometry']].fillna(0)
# Memanggil kolom REMARK, LCODE, GEOMETRY
```

```python
pip install folium matplotlib mapclassify
# Instalasi folium, matplotlib, mapclassify
```

```python
import folium
m = gdf.explore(column="REMARK", name="Banguan")
folium.LayerControl().add_to(m)
m
# Untuk menampilkan peta bangunan
```
